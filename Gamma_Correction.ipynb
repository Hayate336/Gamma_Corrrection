```python
from google.colab.patches import cv2_imshow
import cv2

import matplotlib.pyplot as plt
import numpy as np

#from google.colab import files
#uploaded = files.upload()
```

```python
img = cv2.imread('idk.jpg')
gamma_img = np.array(255*(img/255)**3, dtype = 'uint8')

fig = plt.figure(figsize=(10, 7))

fig.add_subplot(1,2,1)
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(1,2,2)
plt.imshow(gamma_img)
plt.axis('off')
plt.title("Image As Viewed On Screen --> Blacken")
```

```python
img_corr = np.array(255*(img/255)**(1/3), dtype = 'uint8')
gamma_corr = np.array(255*(img_corr/255)**3, dtype = 'uint8')

fig = plt.figure(figsize=(10, 7))

fig.add_subplot(1,3,1)
plt.imshow(img_corr)
plt.axis('off')
plt.title("Gamma Corrected Image")

fig.add_subplot(1,3,2)
plt.imshow(gamma_corr)
plt.axis('off')
plt.title("Image On Screen")

fig.add_subplot(1,3,3)
plt.imshow(img)
plt.axis('off')
plt.title("Original Image (for reference)")
```
